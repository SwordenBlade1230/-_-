```python
import nibabel as nib
import numpy as np
import os
from os.path import join
import time
```

```python
dataset_path = 'kits_test'
```

```python
st = time.time()

print('程式開始執行...')

pic_num = 0

case_name_list = []

for f in os.listdir(dataset_path):
    if 'case_' in f:
        case_name_list.append(f)

for idx, f in enumerate(case_name_list):

    img = nib.load(join(dataset_path, f, 'imaging.nii')).get_fdata()

    print(idx + 210, img.shape)

    pic_num += img.shape[0]

print(pic_num)

ed = time.time()

sp_time = ed - st
print(f'花費時間：{sp_time: .4f}')
```
